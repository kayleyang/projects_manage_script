```python
import time
import requests,json
url = 'http://apm.chtfund.com/graphql'
```

```python
queryServices = {
    "query": "query queryServices($duration: Duration!) {\n    services: getAllServices(duration: $duration) {\n      key: id\n      label: name\n    }\n  }",
    "variables": {
        "duration": {
            "start": "2020-09-01 0000",
            "end": "2020-09-03 2359",
            "step": "MINUTE"
        }
    }
}
queryServicesResponse = requests.post(url,json.dumps(queryServices))
queryServicesResult = json.loads(queryServicesResponse.text)
```

```python
serviceId = '99'
queryEndpoints = {
    "query": "query queryEndpoints($serviceId: ID!, $keyword: String!) {\n    getEndpoints: searchEndpoint(serviceId: $serviceId, keyword: $keyword, limit: 1000) {\n      key: id\n      label: name\n    }\n}",
    "variables": {
        "serviceId": serviceId,
        "keyword": ""
    }
}
```

```python
queryEndpointsResponse = requests.post(url,json.dumps(queryEndpoints))
queryEndpointsResult = json.loads(queryEndpointsResponse.text)
print(queryEndpointsResult)
```

```python
for endpoint in queryEndpointsResult['data']['getEndpoints']:
    endpointName = endpoint['label']
    pageNum = 1
    pageSize = 1000
    while True:
        print(endpointName, ":", pageNum)
        queryTraces = {
            "query": "query queryTraces($condition: TraceQueryCondition) {\n  traces: queryBasicTraces(condition: $condition) {\n    data: traces {\n      key: segmentId\n      endpointNames\n      duration\n      start\n      isError\n      traceIds\n    }\n    total\n  }}",
            "variables": {
                "condition": {
                    "queryDuration": {
                        "start": "2020-09-02 0000",
                        "end": "2020-09-02 2359",
                        "step": "MINUTE"
                    },
                    "traceState": "ALL",
                    "paging": {
                        "pageNum": pageNum,
                        "pageSize": pageSize,
                        "needTotal": True
                    },
                    # "queryOrder": "BY_START_TIME",
                    "queryOrder": "BY_DURATION",
                    "serviceId": serviceId,
                    "endpointName": endpointName
                }
            }
        }

        queryTracesResponse = requests.post(url,json.dumps(queryTraces))
        queryTracesResult = json.loads(queryTracesResponse.text)

        data = queryTracesResult['data']['traces']['data']
        total = queryTracesResult['data']['traces']['total']

        name_list = [time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(int(x['start'])/1000)) for x in data]
        num_list = [x['duration'] for x in data]

        for i in range(len(num_list)):
            if num_list[i] > 1000:
                print(name_list[i], "\t", endpointName, "\t", num_list[i], 'ms')

        if pageNum * pageSize < total:
            pageNum = pageNum + 1
        else:
            break
```
